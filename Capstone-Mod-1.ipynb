```python
# drop unwanted features

features2remove = ['Unnamed: 0.1','Unnamed: 0','id','eiacty','iso3numeric','oil_value_nom','oil_value_2000','oil_value_2014','gas_price_2000',
     'gas_value_nom','gas_value_2000','gas_value_2014','oil_gas_valuePOP_nom','oil_gas_valuePOP_2000','oil_gas_valuePOP_2014',
     'net_oil_exports_mt','net_oil_exports_value','net_oil_exports_valuePOP','net_gas_exports_bcf','net_gas_exports_mboe','net_gas_exports_value','net_gas_exports_valuePOP','mult_2000_2014','net_oil_gas_exports_valuePOP','oil_price_nom'
]

#len(timeSeriesDf['iso3numeric'].unique())
#len(timeSeriesDf['cty_name'].unique())

len(timeSeriesDf['mult_2000_2014'].unique())

cleanedDf = timeSeriesDf.drop(features2remove, axis=1, inplace=False)



for value in cleanedDf:
    print(value+" = ",len(cleanedDf[value].unique()))

```

```python
#one hot encoding

finalDF = pd.get_dummies(cleanedDf, columns=["cty_name"])

finalDF.describe()



```

```python
lable_df = finalDF['oil_price_2000']
features_df = finalDF.drop('oil_price_2000', axis=1, inplace=False)

```
